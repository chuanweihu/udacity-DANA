**Conclusion**

As we can analysis `Guangzhou` and `global` database, we can get the time interval of year between `1840` to `2013`.

```python
# filter assign year
df_global = df_global.query("year >= 1840 & year <=2013")
```

```python
df_global.describe()
```

## 3. Merge Table

```python
# filter "year" and "avg_temp" columns
df1 = df_gz.loc[:, ["year", "avg_temp"]]
```

```python
df1.rename(columns={"avg_temp":"avg_temp_gz"});
```

```python
df2 = df_global.loc[:, ["year", "avg_temp"]]
```

```python
df2.rename(columns={"avg_temp":"avg_temp_global"});
```

```python
# merge two dataframe with the same columns
df = df1.merge(df2, left_on="year", right_on="year", suffixes=("_gz", "_global"))
```

```python
df.info()
```

```python
df.head()
```

## 4. Simple Moving Average

```python
df['SMA_10_gz'] = df.iloc[:, 1].rolling(window=10).mean().round(2)
df['SMA_10_global'] = df.iloc[:, 2].rolling(window=10).mean().round(2)
```

```python
df.head(10)
```

## 5. Line Chart

```python
?plt.subplot
```

```python
x = df['year']
y1 = df['avg_temp_gz']
y2 = df['avg_temp_global']
y3 = df['SMA_10_gz']
y4 = df['SMA_10_global']
```

```python
plt.figure(figsize=[15, 10])

ax1 = plt.subplot(212)
ax1.margins(0.05)
ax1.plot(x, y1, label='Years of Guangzhou')
ax1.plot(x, y2, label='Years of Global')
ax1.plot(x, y1-y2, label="Diff")
ax1.set_title('Years')

ax2 = plt.subplot(221)
ax2.margins(2, 2)
ax2.plot(x, y1-y2)
ax2.set_title('Zoomed Out')

ax3 = plt.subplot(222)
ax3.margins(x=0, y=-0.25)
ax3.plot(x, y1-y2)
ax3.set_title('Zoomed In')

```

```python
plt.figure(figsize=[15, 10])
plt.subplot(211)
plt.plot(x, y1, label='Years of Guangzhou')
plt.plot(x, y2, label='Years of Global')
plt.plot(x, y1-y2, label="Diff")
plt.title('years')
plt.grid(True)
plt.legend();

plt.figure(figsize=[15, 10])
plt.subplot(212)
plt.plot(x, y3, label='SMA 10 Years of Guangzhou')
plt.plot(x, y3-y4, label="Diff")
plt.plot(x, y4, label='SMA 10 Years of Global')
plt.title('SMA 10 years')
plt.grid(True)
plt.legend();
```

Regression Closed Form Solution:
$$\hat{\beta} = (X^{'}X)^{-}X^{'}y$$

```python
import numpy as np
import pandas as pd
import statsmodels.api as sms;
```

```python
df = pd.read_csv('house_prices.csv')
df.head()
```

```python
df['neighborhood'].value_counts()
```

```python
df[['A', 'B', 'C']] =  pd.get_dummies(df.neighborhood)
df.head()
```

```python
df['style'].value_counts()
```

```python
df[['ranch', 'lodge', 'victorian']] = pd.get_dummies(df['style'])
df.head()
```

```python
df['intercept'] = 1

lm = sms.OLS(df['price'], df[['intercept', 'lodge', 'ranch']])
results = lm.fit()
results.summary()
```
